```python
import tensorflow as tf
import tensorflow.keras as k
```

```python
from tensorflow.keras.layers import Conv2D, Add, Input, BatchNormalization, Activation, MaxPooling2D, ZeroPadding2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
```

```python
def nn_model(input_shape):
    X_in = Input(input_shape, name = "input_1")
    X = ZeroPadding2D(padding = (3, 3), name = "conv1_pad")(X_in)
    X = Conv2D(kernel_size = (7, 7), filters = 64, strides = (2, 2), name = "conv1_conv")(X)
    X = BatchNormalization(name = "conv1_bn")(X)
    X = Activation("relu", name = "conv1_relu")(X)
    X = ZeroPadding2D(padding = (1, 1), name = "pool1_pad")(X)
    X = MaxPooling2D((3, 3), strides = (2, 2), name = "pool1_pool")(X)
    X1 = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1))(X)
    X1 = BatchNormalization()(X1)
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 64, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 64, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X1, X])
    X = Activation("relu")(X)
    X2 = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1))(X)
    X2 = BatchNormalization()(X2)
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 64, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 64, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X2, X])
    X = Activation("relu")(X)
    X3 = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1))(X)
    X3 = BatchNormalization()(X3)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 64, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 64, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X3, X])
    X = Activation("relu")(X)
    X4 = Conv2D(kernel_size = (1, 1), filters = 512, strides = (2, 2))(X)
    X4 = BatchNormalization()(X4)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 128, strides = (2, 2))(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 128, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X4, X])
    X = Activation("relu")(X)
    X5 = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1))(X)
    X5 = BatchNormalization()(X5)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 128, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 128, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X5, X])
    X = Activation("relu")(X)
    X6 = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1))(X)
    X6 = BatchNormalization()(X6)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 128, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 128, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X6, X])
    X = Activation("relu")(X)
    X7 = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1))(X)
    X7 = BatchNormalization()(X7)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 128, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 128, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X7, X])
    X = Activation("relu")(X)
    X8 = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (2, 2))(X)
    X8 = BatchNormalization()(X8)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (2, 2))(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X8, X])
    X = Activation("relu")(X)
    X9 = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1))(X)
    X9 = BatchNormalization()(X9)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    x = Add()([X9, X])
    X = Activation("relu")(X)
    X10 = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1))(X)
    X10 = BatchNormalization()(X10)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X10, X])
    X = Activation("relu")(X)
    X11 = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1))(X)
    X11 = BatchNormalization()(X11)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X11, X])
    X = Activation("relu")(X)
    X12 = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1))(X)
    X12 = BatchNormalization()(X12)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X12, X])
    X = Activation("relu")(X)
    X13 = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1))(X)
    X13 = BatchNormalization()(X13)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 256, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 1024, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X13, X])
    X = Activation("relu")(X)
    X14 = Conv2D(kernel_size = (1, 1), filters = 2048, strides = (2, 2))(X)
    X14 = BatchNormalization()(X14)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 512, strides = (2, 2))(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 2048, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X14, X])
    X = Activation("relu")(X)
    X15 = Conv2D(kernel_size = (1, 1), filters = 2048, strides = (1, 1))(X)
    X15 = BatchNormalization()(X15)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 2048, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X15, X])
    X = Activation("relu")(X)
    X16 = Conv2D(kernel_size = (1, 1), filters = 2048, strides = (1, 1))(X)
    X16 = BatchNormalization()(X16)
    
    
    
    X = Conv2D(kernel_size = (1, 1), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (3, 3), filters = 512, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Activation("relu")(X)
    
    X = Conv2D(kernel_size = (1, 1), filters = 2048, strides = (1, 1), padding = "same")(X)
    X = BatchNormalization()(X)
    X = Add()([X16, X])
    X = Activation("relu")(X)
    
    model = Model(inputs = X_in, outputs = X)
    
    return model
    
```
